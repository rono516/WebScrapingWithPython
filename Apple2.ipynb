```python
#Read in the data
df = pd.read_csv('AAPL.csv')
#Show data
df
```

```python
#Select the date and the price
df = df[['Date', 'Close']]
#Rename the features:
df = df.rename(columns={'Date': 'ds', 'Close':'y'})
# Show the data
df
```

```python
#Get last 20 rows of data and store into new variable
last= df[len(df)-20:]
last
```

```python
# Get all of the rows in the dataset except the last 20 rows of data
df = df[:-20]
df
```

```python
#import facebook's prophet library
from fbprophet import Prophet
#Create prophet object(model)
fbp = Prophet(daily_seasonality = True)

#fit or train the model
fbp.fit(df)
future = fbp.make_future_dataframe(periods= 365)
forecast = fbp.predict(future)
```

```python
#import the facebook prophet plotting libraries
from fbprophet.plot import plot_plotly

#plot the data
plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for 2021-12-06
forecast[forecast.ds == '2021-12-06']['yhat']
```

```python
#Show the actual value of the stock for 2021-12-06
last[last.ds == '2021-12-06']['y']
```
